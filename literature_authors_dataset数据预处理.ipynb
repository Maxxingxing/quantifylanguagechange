```python
#生成总体样本，将作者作为一个variable
import os
import sys
import pandas as pd
import string
for filename in os.listdir("DCLSA/literature_author_dataset"):
    list = filename.split('.')
    author = list[0]
    print(author)
    df = pd.read_csv("DCLSA/literature_author_dataset/"+filename)
    df['author']=author
    print(df.head())
    #print(type(df))
    if os.path.exists("all.csv"):
        df.to_csv("all.csv",mode='a',index=0,header=False)
    else:
        df.to_csv("all.csv",mode='a',index=0)
```

```python
#全集和中author变量的编码化
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

#df1.drop(['Unnamed: 0'], axis=1,inplace=True)


#离散变量one-hot-code
#方法见————https://blog.csdn.net/wotui1842/article/details/80697444

le = LabelEncoder()
dis_author = le.fit_transform(df1['author'].values)
#     print(dis_author)
#     print(type(dis_author))
dis_author = dis_author.reshape(-1, 1)
a_dis = dis_author.tolist()
enc = OneHotEncoder()
enc.fit(a_dis)
b_dis = enc.transform(a_dis).toarray()
#print(b_dis)
#b_dis = list(b_dis)
#print(type(b_dis))
df2 = pd.DataFrame(b_dis) 
#len(df2)

df3 = pd.concat([df1,df2],axis=1)
df3.head()

df3.to_csv("all_ohc.csv",index=0)
```
